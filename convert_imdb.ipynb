```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
from tqdm import tqdm
import pickle
```

```python
path = "Models/newImdb/imdbdoc.mat"

# Usage
with h5py.File(path, "r") as f:
    print(list(f.keys()))
    data = f["imdb"][()]
    a = [[f[data[i][j]][()] for j in range(4)] for i in tqdm(range(len(data)))]
```

```python
with open('imdb_doc.pkl', 'wb') as f:
    pickle.dump(a, f)
```

```python
# 1: train, 2: val
np.unique(np.array([x[1] for x in a]).squeeze(), return_counts=True)
```

```python
# index 0: rgb
# index 1: dataset split
# index 2: angle
# index 3: contours
plt.imshow(a[0][0].T / np.max(a[0][0]), cmap='gray')
plt.show()
plt.imshow(a[0][3].T / np.max(a[0][0]), cmap='gray')
plt.show()
```
